##### Build the decision tree model using the following steps:
1. Create two subsets of data  one for training data and the other for test data. For this activity, set an appropriate seed value and use 78 instances/rows out of the 148 instances/rows for training data and the remaining 70 instances/rows for testing data.
2. Train the decision tree.


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn import metrics
```

Split the data into dependent and independent datasets 

```python
feature_cols = ["lymphatics", "early_uptake", "lymnodes_enlar",	"lym_changes",	"node_defect", "node_changes", "stru_changes", "special_forms", "no_of_nodes"]
X = df[feature_cols] 
y = df['class']
X.head()
```

```python
y.head()
```

Set up the test and training datasets .... 

Set-up the Decision Tree Classifier: 

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y.values)
y1 = le.transform(y.values)
print(y1)
```

```python
le.fit(X['lymphatics'].values)
X['lymphatics'] = le.transform(X['lymphatics'].values)
print(X['lymphatics'])
```

```python
le.fit(X.loc[:,'early_uptake'].values)
X.loc[: ,'early_uptake'] = le.transform(X.loc[:,'early_uptake'].values)
print(X['early_uptake'])
```

```python
le.fit(X.loc[:,'lym_changes'].values)
X.loc[: ,'lym_changes'] = le.transform(X.loc[:,'lym_changes'].values)
print(X['lym_changes'])
```

```python
le.fit(X['node_defect'].values)
X['node_defect'] = le.transform(X['node_defect'].values)
print(X['node_defect'])
```

```python
le.fit(X['node_changes'].values)
X['node_changes'] = le.transform(X['node_changes'].values)
print(X['node_changes'])
```

```python
le.fit(X['stru_changes'].values)
X['stru_changes'] = le.transform(X['stru_changes'].values)
print(X['stru_changes'])
```

```python
le.fit(X['special_forms'].values)
X['special_forms'] = le.transform(X['special_forms'].values)
print(X['special_forms'])
```

```python
le.fit(X['no_of_nodes'].values)
X['no_of_nodes'] = le.transform(X['no_of_nodes'].values)
print(X['no_of_nodes'])
```

```python
X.head(10)
```

```python
test_size = (70/148)
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size=test_size, random_state=1)
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
print('Accuracy on training dataset: {:.3f}'. format(tree.score(X_train, y_train)))
print('Accuracy on test dataset: {:.3f}'. format(tree.score(X_test, y_test)))
```

Accuracy on training dataset very close to 1.000, as expected. 
Test dataset accuracy is very moderate - overfitting??? 

Limit the number of decision nodes on the tree...... 

```python
n=1
for n in range(1, 9):
    tree = DecisionTreeClassifier(max_depth = (n), random_state =1)
    tree.fit(X_train, y_train)
    print(f'Depth of tree : {n}' )
    print('Accuracy on training dataset: {:.3f}'. format(tree.score(X_train, y_train)))
    print('Accuracy on test dataset: {:.3f}'. format(tree.score(X_test, y_test)))
```

The optimum depth of tree is 3 decisions for this dataset.  

The hyperparameters can be easier to find using the GridSearchCV function in sklearn .....

```python
from sklearn.model_selection import GridSearchCV

params = {
    'criterion':  ['gini', 'entropy'],
    'max_depth':  [2, 3, 4, 5, 6, 7],
    'max_features': [None, 'sqrt', 'log2', 0.2, 0.4, 0.6, 0.8],
    'splitter': ['best', 'random']
}

clf = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=params,
    cv=5,
    n_jobs=5,
    verbose=1,
)

clf.fit(X_train, y_train)
print(clf.best_params_)
```

For this decision tree - the best hyperparameters are - criterion: gini, max_depth: 3, max_features: 0.8, splitter: best

Next, we can have a look at what this tree looks like ......

```python
best_tree = DecisionTreeClassifier(criterion = 'gini', max_depth = 3, max_features = 0.8, splitter = 'best', random_state =1)
best_tree.fit(X_train, y_train)
```
