#### b) Find the singular value decomposition (SVD) of matrix A.

For a matrix $A \in \mathbb{R}^{mxm}$ of rank $r \in [0, min(m,n)]$. The SVD of A is a decomposition of the form:

$$ A= U \Sigma V^T $$ 

* Where $U \in \mathbb{R}^{mxm}$ is an orthogonal matrix with column vector $u_{i} (i = 1, ... , m)$.

    * Matrix U diagonalizes $AA^T$

    * $u_{i}'$ s are called left-singular vectors and are eigenvectors of $AA^T$.

* $V \in \mathbb{R}^{nxn}$ is an orthogonal matrix with column vector $v_i (i = 1, ... , n)$.

    * Matrix V diagonalizes $A^TA$.

    * $v_{i}'$ s are called left-singular vectors and are eigenvectors of $A^TA$.

* $\Sigma \in \mathbb{R}^{mxn}$ is a diagonal matrix with $\Sigma_{ii} = \sigma_i \geq 0$ and  $\Sigma_{ij} = 0. $
$\sigma_i$ is called the singular values of A. 

    * $\sigma_i$ is the square roots of the eigenvalues of $A^TA$.

Calculating $AA^T$ :

$\begin{vmatrix} 2 & -3 \\ 1 & 6 \end{vmatrix} \cdot \begin{vmatrix} 2 & 1 \\ -3 & 6 \end{vmatrix}$

Eigenvalues are the roots of the characteristic equation $(A - \lambda I) = 0 $

```python
#### Checking SVD using the code given in Week 4 :

# This task is to practice singular value decomposition (SVD)

import numpy as np
from scipy import linalg

print('Checking SVD using the code given in Week 4 :')

# 2 x 3 matrix
A = np.array([[2, -3],
               [1, 6]])

U, Sigma, V_T = np.linalg.svd(A)

print(f'Left orthogonal matrix for A:\n{np.round(U, 2)}\n')

# Display singular values along the diagonal of sigma matrix.
print(f'Singular values diagonal matrix for A:\n{np.round(Sigma, 2)}\n')

print(f'Right orthogonal matrix for A:\n{np.round(V_T, 2)}') 
```
